Load the Required Libraries, env and create open AI client

```python
from openai import OpenAI
from dotenv import load_dotenv
import requests
from PIL import Image
from io import BytesIO
 
load_dotenv()
client = OpenAI()
```

Define a function that would take in a prompt and generate a URL

```python
def generate_image(user_prompt):
    response = client.images.generate(
        model="dall-e-3",
        prompt=user_prompt,
        size="1024x1024",
        quality="standard",
        n=1,
    )
    image_url = response.data[0].url
    image_data = requests.get(image_url).content
    return Image.open(BytesIO(image_data))
```

Import Gradio and generate an image

```python
import gradio as gr

myscreen = gr.Interface(
    fn=generate_image,
    inputs=gr.Textbox(label="Enter the prompt for image generation"),
    outputs=gr.Image(label="Here is your image"),
    flagging_mode="never"
)

myscreen.launch()
```
